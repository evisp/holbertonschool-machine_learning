### 4. Hello, derivatives!

$$\frac{dy}{dx}$$ where $$y = x^4 + 3x^3 - 5x + 1$$

To find the derivative of y with respect to x, we need to differentiate each term of y with respect to x using the power rule of derivatives:

$$\frac{dy}{dx} = \frac{d}{dx}(x^4) + \frac{d}{dx}(3x^3) - \frac{d}{dx}(5x) + \frac{d}{dx}(1)$$

$$\frac{dy}{dx} = 4x^3 + 9x^2 - 5$$

Therefore, the derivative of `y` with respect to `x` is $$4x^3 + 9x^2 - 5$$.

### 5. A log on the fire

$$
\frac{d (xln(x))}{dx}
$$

To solve this derivative, we can use the product rule of differentiation:

Let `u = x` and `v = ln(x)`

Then, we have:

$$\frac{du}{dx} = 1$$

$$\frac{dv}{dx} = \frac{1}{x}$$

Using the product rule, we get:

$$
\frac{d}{dx} (x\cdot ln(x)) = u\frac{dv}{dx} + v\frac{du}{dx}
= x\frac{1}{x} + ln(x)\cdot(1)
= 1 + ln(x)
$$

Therefore, the derivative of `x ln(x)` with respect to `x` is `1 + ln(x)`.





### 6. It is difficult to free fools from the chains they revere

$$\frac{d (ln(x^2))}{dx}$$

Using the chain rule and power rule, we can solve this derivative as follows:

\begin{aligned}
\frac{d}{dx} \ln(x^2) &= \frac{1}{x^2}\cdot\frac{d}{dx}(x^2)\
&= \frac{1}{x^2} \cdot 2x\
&= \frac{2}{x}
\end{aligned}

Therefore, the derivative of $ln(x^2)$ with respect to $x$ is $\frac{2}{x}$.

### 7. Partial truths are often more insidious than total falsehoods

$$\frac{\partial f(x, y)}{\partial y}$$ 

where 

$$f(x, y) = e^{xy}$$ 

and 

$$\frac{\partial x}{\partial y}=\frac{\partial y}{\partial x}=0$$

To solve this derivative, we use the chain rule of differentiation. We have:

- Write out the derivative using the chain rule:
$$\frac{\partial f(x, y)}{\partial y} = \frac{\partial}{\partial y}(e^{xy}) = \frac{\partial}{\partial y}(u) = \frac{\partial u}{\partial v}\frac{\partial v}{\partial y},$$
where $u=e^{xy}$ and $v=xy$.


- Find $\frac{\partial u}{\partial v}$ by taking the derivative of $e^v$ with respect to $v$:
$$\frac{\partial u}{\partial v} = \frac{\partial}{\partial v}(e^{xy}) = e^{xy}.$$


- Find $\frac{\partial v}{\partial y}$ by using the product rule:
$$\frac{\partial v}{\partial y} = \frac{\partial}{\partial y}(xy) = x\frac{\partial y}{\partial y} + y\frac{\partial x}{\partial y} = x\cdot 1 + y\cdot 0 = x.$$


- Substitute the values of $\frac{\partial u}{\partial v}$ and $\frac{\partial v}{\partial y}$ into the derivative from step 1:
$$\frac{\partial f(x, y)}{\partial y} = \frac{\partial u}{\partial v}\frac{\partial v}{\partial y} = e^{xy}\cdot x = x\cdot e^{xy}.$$


Therefore, the partial derivative of $f(x, y) = e^{xy}$ with respect to $y$, given that $\frac{\partial x}{\partial y} = \frac{\partial y}{\partial x} = 0$, is $x\cdot e^{xy}$.







### 8. Put it all together and what do you get?

$$
\frac{\partial^2}{\partial y\partial x}(e^{x^2y})
$$

where 

$$
\frac{\partial x}{\partial y}=\frac{\partial y}{\partial x}=0
$$

To compute the second partial derivative of $e^{x^2y}$ with respect to $y$ and then $x$, we can use the following steps:

Step 1: Take the partial derivative of the function with respect to $y$ while treating $x$ as a constant. This gives:

$$\frac{\partial}{\partial y}(e^{x^2y}) = x^2 e^{x^2y}$$

Step 2: Take the partial derivative of the result obtained in Step 1 with respect to $x$ while treating $y$ as a constant. This gives:

$$\frac{\partial}{\partial x}\left(\frac{\partial}{\partial y}(e^{x^2y})\right) = \frac{\partial}{\partial x}(x^2 e^{x^2y})$$

Step 3: Apply the product rule to the expression in Step 2. This gives:

$$\frac{\partial}{\partial x}(x^2 e^{x^2y}) = \frac{\partial x^2}{\partial x} e^{x^2y} + x^2 \frac{\partial}{\partial x}(e^{x^2y})$$

Step 4: Simplify the first term using the constant rule, which gives:

$$\frac{\partial x^2}{\partial x} e^{x^2y} = 2xe^{x^2y}$$

Step 5: Take the partial derivative of $e^{x^2y}$ with respect to $x$ using the chain rule. This gives:

$$\frac{\partial}{\partial x}(e^{x^2y}) = 2xye^{x^2y}$$

Step 6: Substitute the results obtained in Steps 4 and 5 back into the expression in Step 3. This gives:

$$\frac{\partial}{\partial x}\left(\frac{\partial}{\partial y}(e^{x^2y})\right) = 2xe^{x^2y} + x^2(2xye^{x^2y})$$

Step 7: Simplify the expression obtained in Step 6 by factoring out the common factor of $2xe^{x^2y}$. This gives:

$$\frac{\partial^2}{\partial y\partial x}(e^{x^2y}) = 2xe^{x^2y} + 2x^3ye^{x^2y}$$

Thus, the second partial derivative of $e^{x^2y}$ with respect to $y$ and then $x$ is equal to $2xe^{x^2y} + 2x^3ye^{x^2y} = 2x(1 + x^2y)e^{x^2y}$.

To solve the integral $\int x^3 dx$, we can use the power rule of integration which states that for any constant $n \neq -1$,

$$
\int x^n dx = \frac{x^{n+1}}{n+1} + C
$$

where $C$ is the constant of integration. Using this formula, we have

$$
\int x^3 dx = \frac{x^{3+1}}{3+1} + C = \frac{x^{4}}{4} + C 
$$

$$
\int e^{u} du =e^u + C = e^{2y} + C

Substituting this result back into the original equation, we have:

$$
\int e^{2y} dy = \frac{1}{2} \int e^u du = \frac{1}{2} e^{2y}+ C

### 15. Winners are people with definite purpose in life

$$
\int_{0}^{5} x dy
$$

We are given the integral:

$$\int_{0}^{5} x \ dy$$

To evaluate this integral with respect to $y$, we can treat $x$ as a constant and use the power rule of integration:

$$\int x \ dy = xy + C$$

where $C$ is the constant of integration.

Since we are evaluating the integral from $y=0$ to $y=5$, we can use the fundamental theorem of calculus to find the definite integral:

$$\int_{0}^{5} x \ dy = \left[xy\right]_{0}^{5} = x(5) - x(0) = 5x$$

Therefore, the solution to the integral $\int_{0}^{5} x \ dy$ with respect to $y$ is $5x$.

### 16. Double whammy

$$
\int_{1}^{2} \int_{0}^{3} x^2 y^{-1} dx dy
$$

We first integrate with respect to $x$, treating $y$ as a constant, which gives us:

$$\int_{0}^{3} x^2 y^{-1} dx = y^{-1} \frac{x^3}{3} \bigg\rvert_{0}^{3} = \frac{3^3}{3y} - \frac{0^3}{3y} = \frac{9}{y}$$

We then integrate with respect to $y$, which gives us:

$$\int_{1}^{2} \frac{9}{y} dy = 9 \ln(y) \bigg\rvert_{1}^{2} = 9 \ln(2)$$

Therefore, the value of the double integral is $9\ln(2)$.

## Visualize Derivatives and Integrals

```python
# start by importing the necessary libraries
import matplotlib.pyplot as plt
import numpy as np
```

### Visualize derivatives

#### Function $$y = x^4 + 3x^3 - 5x + 1$$

```python
# Define the function and its derivative
def f(x):
    return x**4 + 3*x**3 - 5*x + 1

def dfdx(x):
    return 4*x**3 + 9*x**2 - 5

# Create a range of x values
x = np.linspace(-3, 2, 100)

# Evaluate the function and its derivative at each x value
y = f(x)
dydx = dfdx(x)

# Create a plot of the function and its derivative
fig, ax = plt.subplots()
ax.plot(x, y, label='y = x^4 + 3x^3 - 5x + 1')
ax.plot(x, dydx, label="dy/dx")
ax.axhline(y=0, c='black', linestyle='--')
ax.axvline(x=0, c='black', linestyle='--')
ax.legend()
plt.show()
```

#### Function $$\frac{d (ln(x^2))}{dx}$$

```python
# Define the function and its derivative
def f(x):
    return np.log(x**2)

def dfdx(x):
    return 2/x

# Create a range of x values
x = np.linspace(0.1, 3, 100)

# Evaluate the function and its derivative at each x value
y = f(x)
dydx = dfdx(x)

# Create a plot of the function and its derivative
fig, ax = plt.subplots()
ax.plot(x, y, label='y = d(ln(x^2))/dx')
ax.plot(x, dydx, label="dy/dx")
ax.axhline(y=0, color='k', linestyle='--')
ax.axvline(x=0, color='k', linestyle='--')
ax.legend()
plt.show()
```

### Visualize integrals

#### Function $$y = u^2$$

```python
# define the function to be integrated
def f(u):
    return u**2

# Create an array of values for the independent variable u
u = np.linspace(0, 3, 1000)

# Evaluate the function f at each value of u
y = f(u)

"""
Use matplotlib to plot the area under the curve defined by f from 0 to 3 
using the fill_between function. 
Set the x and y limits of the plot to match the limits of the integral
"""
plt.plot(u, y)
plt.fill_between(u, y, 0, where=(u>=0)&(u<=3), color='orange', alpha=0.5)
plt.xlim(0, 3)
plt.ylim(0, 9)
plt.xlabel('u')
plt.ylabel('u^2')
plt.title('Area under the curve of u^2 from 0 to 3')
plt.show()
```

#### Function 

$$y = \frac{x^2}{y}$$

```python

# define the function to be integrated
def f(x, y):
    return x**2 / y

# Create a grid of x and y values using numpy.meshgrid
x = np.linspace(1, 2, 50)
y = np.linspace(0.1, 3, 50)
X, Y = np.meshgrid(x, y)

# Evaluate the function f at each combination of x and y values
Z = f(X, Y)

"""
Use matplotlib to create a 3D plot of the volume defined by the function f over the specified ranges. 
Use the plot_surface function to create a surface plot of the volume, 
and set the colormap to 'viridis' for better visibility
"""

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
plt.title('Volume under the surface defined by f(x,y)')
plt.show()
```

### Visualize partial second order derivatives

#### Function

$$\frac{\partial^2}{\partial y\partial x}(e^{x^2y})$$

whose derivative is 

$$
2xe^{x^2y} + 2x^3ye^{x^2y}
$$

```python
# Define the function and its second partial derivative
def f(x, y):
    return np.exp(x**2 * y)

def d2fdxdy(x, y):
    return 2 * x * np.exp(x**2 * y) + 2 * x**3 * y * np.exp(x**2 * y)

# Create a range of x and y values
x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
X, Y = np.meshgrid(x, y)

# Evaluate the function and its second partial derivative at each (x,y) pair
Z = f(X, Y)
d2Zdxdy = d2fdxdy(X, Y)

# Create a 3D plot of the function and its second partial derivative
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x,y)')
ax.set_title('Function f(x,y) = e^(x^2y)')

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, d2Zdxdy, cmap='viridis', edgecolor='none')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('d^2f/dxdy')
ax.set_title('Second partial derivative of f(x,y)')

plt.show()
```
