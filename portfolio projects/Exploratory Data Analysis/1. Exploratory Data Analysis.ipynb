### Iris dataset: simple, but effective

Iris is a classic dataset with measurements of iris flowers, suitable for understanding basic EDA concepts
- The Iris dataset is a classic dataset in machine learning and statistics.
- It contains measurements of sepals and petals of three different species of iris flowers: Setosa, Versicolor, and Virginica.
- Each sample consists of four features: sepal length, sepal width, petal length, and petal width.
- The dataset contains a total of 150 samples, with 50 samples for each species.

To get you started, here's Python code to load the Iris dataset using the `load_iris` function from the `sklearn.datasets` module:

```python
from sklearn.datasets import load_iris

# load the dataset
iris_data = load_iris()

# Extract features and target variable
X = iris_data.data    # Features (sepal length, sepal width, petal length, petal width)
y = iris_data.target  # Target variable (species)

# Print the shape of the dataset
print("Shape of features:", X.shape)
print("Shape of target variable:", y.shape)
```

### California housing dataset

A real-world dataset containing housing data in California, providing a more complex and diverse set of features for exploration.

- The California Housing dataset contains housing data for various districts in California.
- It includes features such as median house value, median income, housing median age, total rooms, total bedrooms, population, households, latitude, and longitude.
- The target variable is the median house value for California districts.
- It consists of 20,640 samples, each representing a district in California

```python
from sklearn.datasets import fetch_california_housing

# Load the California Housing dataset
california_housing = fetch_california_housing()

# Extract features and target variable
X_california = california_housing.data    # Features
y_california = california_housing.target  # Target variable (median house value)


print("Shape of features:", X_california.shape)
print("Shape of target variable:", y_california.shape)
```

### The Titanic dataset

Ever wondered if Jack could be saved? Explore the Titanic dataset and sink in the world of data. The dataset provides information on passengers, their demographics, and survival status, making it suitable for classification and survival analysis

- The Titanic dataset contains information about passengers aboard the RMS Titanic, including whether they survived or not.
- It includes features such as passenger class, sex, age, number of siblings/spouses aboard, number of parents/children aboard, ticket fare, cabin number, and port of embarkation.
- The target variable is whether a passenger survived the Titanic disaster (1 for survived, 0 for not survived).
- It consists of 891 samples, each representing a passenger on the Titanic

```python
import seaborn as sns

# Load the Titanic dataset
titanic_data = sns.load_dataset('titanic')

# Extract features and target variable
X_titanic = titanic_data.drop(columns=['survived'])  # Features
y_titanic = titanic_data['survived']                 # Target variable (survived)

print("Shape of features:", X_titanic.shape)
print("Shape of target variable:", y_titanic.shape)
```

### Being happy fan? 

Explore factors influencing happiness scores across countries. The World Happiness Report is a landmark survey of the state of global happiness.

The happiness scores and rankings use data from the Gallup World Poll. The scores are based on answers to the main life evaluation question asked in the poll. This question, known as the Cantril ladder, asks respondents to think of a ladder with the best possible life for them being a 10 and the worst possible life being a 0 and to rate their own current lives on that scale. The scores are from nationally representative samples for the years 2013-2016 and use the Gallup weights to make the estimates representative.

You can get the dataset [here](https://www.kaggle.com/datasets/unsdsn/world-happiness)
