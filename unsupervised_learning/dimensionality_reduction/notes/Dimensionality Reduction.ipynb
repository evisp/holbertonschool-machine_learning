# The Big Picture: Machine Learning Categories

**Machine Learning** is a field of artificial intelligence focused on developing algorithms and models that **enable computers to learn from data and make predictions** or decisions without being explicitly programmed.

<img src="../../figs/mlcategories.jpg" alt="logo" width="600"/>


Three Different Categories of Machine Learning:

## Supervised Learning:

- Involves labeled training data (input-output pairs).


- Goal is to learn a mapping from inputs to outputs.


- Common applications include classification and regression tasks.


- Examples include spam email detection, image recognition, and predicting house prices.


## Unsupervised Learning:

- Deals with unlabeled data, finding patterns or structures within it.


- No predefined target or output.


- Common techniques include clustering and dimensionality reduction.


- Examples include customer segmentation and topic modeling in text data.


## Reinforcement Learning:

- Focuses on training agents to make sequences of decisions.


- Agents learn through trial and error by interacting with an environment.


- Goal is to maximize a cumulative reward signal.


- Applications range from game-playing AI (e.g., AlphaGo) to autonomous robotics.

# Dimensionality Reduction

In data analysis and machine learning, we often deal with **datasets that have a large number of features or dimensions**. While having many features can provide valuable information, it can also lead to **challenges such as increased computational complexity, overfitting, and difficulties in visualization**. 

**Dimensionality Reduction** techniques help address these issues by **reducing the number of features while preserving as much relevant information as possible**.



In this notebook, we will explore the concept of Dimensionality Reduction, its importance, and two popular techniques: `Principal Component Analysis (PCA)` and `t-Distributed Stochastic Neighbor Embedding (t-SNE)`. 


## What is dimensionality reduction and what are its purposes?

### Curse of Dimensionality

**The Curse of Dimensionality refers to the exponential increase in data volume as the number of dimensions or features increases**. This leads to several problems:


- **Increased Computational Complexity**: More dimensions require more computational resources and time to process data.


- **Overfitting**: High-dimensional data can lead to overfitting, where models perform well on training data but poorly on unseen data.


- **Visualization Challenges**: Visualizing data beyond 3 dimensions is difficult, making it hard to explore and understand high-dimensional data.

<img src="../../figs/dimensionality_reduction.jpeg" alt="logo" width="800"/>



### Benefits of Dimensionality Reduction

Dimensionality Reduction techniques offer the following advantages:

- **Simplification**: Reduce the complexity of data, making it easier to work with.


- **Faster Processing**: Less data to process leads to quicker model training and evaluation.


- **Improved Visualization**: Lower-dimensional data can be visualized more easily.


- **Reduced Noise**: Eliminate irrelevant features, reducing the impact of noisy data.

## What is principal components analysis (PCA)?

### Understanding PCA

Principal Component Analysis (PCA) is a linear Dimensionality Reduction technique that finds new orthogonal axes (principal components) along which the data has the maximum variance. The first principal component explains the most variance, the second explains the second most, and so on.

### PCA in Simple English

PCA is like finding the best way to look at your data from a different angle, where you can see the most important things clearly.

Imagine you have a scatterplot with lots of data points spread out in many directions. These data points represent your features (like height, weight, age, etc.) for each item in your dataset. You might be wondering, "How can I make sense of all this data and find what matters the most?"

PCA helps you with this by giving you a new set of axes, or directions, to look at your data. These new axes are special because they are chosen in a way that the first one points in the direction where your data varies the most. In other words, it points to the line that captures the most important information.

<img src="../../figs/pca.png" alt="logo" width="800"/>


The second axis, which is also new and different from the first, points in the direction where your data varies the second most. It's like finding the next most important information after the first axis has already explained as much as it can.

In technical terms, these new axes are called "principal components." The first principal component explains the most variance, which means it tells you the most about your data. The second principal component explains the second most variance, and so on.

### PCA in Technical Terms

PCA finds the principal components through a mathematical process that involves linear transformations. 


- **Centering the Data**: PCA starts by centering the data, which means subtracting the mean (average) value from each data point. This step ensures that the data is centered around the origin of the coordinate system.


- **Calculating the Covariance Matrix**: Next, it calculates the covariance matrix. This matrix measures how variables change together. It shows which variables are positively or negatively correlated.


- **Eigenvalue Decomposition**: PCA then performs eigenvalue decomposition on the covariance matrix. This process finds the eigenvalues and eigenvectors of the matrix.Eigenvalues tell us how much each principal component matters (how much variance it captures), and eigenvectors tell us the directions in which the data varies the most (the directions of the principal components)
    - *Eigenvalues* are scalar values that represent how a linear transformation (like the one applied in PCA) stretches or compresses data along specific directions. They indicate how much variance is captured by each principal component in PCA. Larger eigenvalues correspond to more important directions.
    - *Eigenvectors* are non-zero vectors that point in the directions along which a linear transformation only stretches or compresses data, without changing their orientation. In PCA, eigenvectors represent the principal components. Each eigenvector corresponds to a specific eigenvalue and defines a direction in the original data space


- **Selecting Principal Components**: The eigenvectors are the principal components. They represent the directions in the original data space along which the data varies the most. The corresponding eigenvalues indicate the variance of the data along those directions.


- **Sorting Principal Components**: The principal components are usually sorted in descending order of their eigenvalues. The first principal component (PC1) explains the most variance in the data, the second principal component (PC2) explains the second most, and so on.


- **Reducing Dimensionality**: To reduce the dimensionality of the data, you can choose to keep only the top N principal components that capture most of the variance. This simplifies the data while retaining most of the important information.

## What is the difference between linear and non-linear dimensionality reduction?

The primary difference between linear and non-linear dimensionality reduction techniques lies in **how they model the relationships between data points and reduce the dimensionality of the dataset**:


### Linear Dimensionality Reduction:

- Linear methods, like `Principal Component Analysis (PCA)` and `Linear Discriminant Analysis (LDA)`, assume that the relationships between features in the dataset are linear.


- They aim to find linear combinations of the original features that capture the most variance or discriminative information.


- Linear dimensionality reduction techniques are effective when the data can be well-approximated by linear transformations and when preserving global structures, such as variance or class separation, is essential.


- These methods are computationally efficient and relatively straightforward to interpret.


### Non-linear Dimensionality Reduction:

- Non-linear methods, like `t-Distributed Stochastic Neighbor Embedding (t-SNE)` and `Isomap`, do not assume linearity in the data relationships.


- They focus on preserving local relationships and capturing complex, non-linear structures in the data.


- Non-linear dimensionality reduction is beneficial when data exhibits intricate patterns or clusters that cannot be adequately represented using linear techniques.


- These methods often involve more complex computations and may be less intuitive to interpret than linear techniques.

## Which techniques are linear/non-linear?

In practical terms, the choice between linear and non-linear dimensionality reduction methods **should be guided by the nature of your data** and your specific goals. 


If your data exhibits clear linear relationships, linear techniques like Principal Component Analysis (PCA) are appropriate, especially when reducing computational complexity or capturing global variance is essential. 


On the other hand, if your data contains complex non-linear structures, consider non-linear techniques like t-Distributed Stochastic Neighbor Embedding (t-SNE) for visualization or Isomap for manifold learning. Evaluate and compare the methods based on your objectives, keeping in mind that non-linear techniques may be computationally expensive and may not always generalize well, while linear techniques are typically more interpretable but may not capture complex data structures. 


Domain knowledge and experimentation are valuable in making the final choice.


### Linear Dimensionality Reduction Techniques:

1. `Principal Component Analysis (PCA)`: Linear technique that finds orthogonal axes capturing maximum variance.


2. `Linear Discriminant Analysis (LDA)`: Linear method used for supervised dimensionality reduction, emphasizing class separability.


3. `Factor Analysis (FA)`: Linear technique that models data as a linear combination of factors and error terms.


### Non-linear Dimensionality Reduction Techniques:

1. `t-Distributed Stochastic Neighbor Embedding (t-SNE)`: Non-linear technique for visualization, emphasizing local similarity preservation.


2. `Isomap`: Non-linear technique that uncovers data's intrinsic manifold structure based on geodesic distances.


3. `Locally Linear Embedding (LLE)`: Non-linear method preserving local relationships by approximating each data point as a linear combination of its neighbors.


4. `Kernel PCA`: Extension of PCA using kernel tricks to capture non-linear patterns by mapping data to a higher-dimensional space.


5. `Autoencoders (Neural Networks)`: Non-linear neural network architectures used for unsupervised feature learning and dimensionality reduction.


6. `Multi-Dimensional Scaling (MDS)`: Non-linear technique preserving pairwise distances between data points in a lower-dimensional space.





## What is t-distributed stochastic neighbor embedding (t-SNE)?

`t-Distributed Stochastic Neighbor Embedding (t-SNE)` is a **non-linear dimensionality reduction technique commonly used for data visualization and exploration**. It was introduced by Laurens van der Maaten and Geoffrey Hinton in 2008. t-SNE is particularly effective at revealing the **underlying structure and patterns in high-dimensional data by mapping it to a lower-dimensional space** while preserving local similarities between data points. Here's how t-SNE works:

- **Local Similarities Preservation**: t-SNE starts by measuring the pairwise similarities between data points in the high-dimensional space. These similarities are typically based on the probability distribution of pairwise distances. Data points that are similar in the high-dimensional space will have higher probabilities of being selected as neighbors.


- **Mapping to Lower-Dimensional Space**: Next, t-SNE constructs a similar probability distribution in a lower-dimensional space (usually 2D or 3D). It aims to find a mapping of the data points in the lower-dimensional space such that similar data points in the original space remain close together while dissimilar points are pushed apart. It does this by minimizing the divergence between the two probability distributions, using gradient descent optimization.


- **Heavy-Tailed t-Distribution**: It is a probability distribution used to model pairwise similarities between data points in both the original high-dimensional space and the lower-dimensional space created during the dimensionality reduction process. This distribution is "heavy-tailed," meaning it assigns relatively high probabilities to capturing both very similar (close) and somewhat dissimilar (not-so-close) relationships between data points. This heavy-tailed property is beneficial because it prevents points that are far apart in the high-dimensional space from being too densely packed together in the lower-dimensional representation. It helps maintain a more faithful representation of the data's structure, especially when dealing with complex, non-linear relationships among data points.


- **Iterative Optimization**: t-SNE is an iterative optimization process. It gradually adjusts the positions of data points in the lower-dimensional space to find the configuration that minimizes the divergence between the two probability distributions.


- **Parameter Tuning**: t-SNE has a few important hyperparameters, such as the perplexity (controlling the balance between preserving local and global structure) and the learning rate. Proper parameter tuning is crucial to the success of the technique.


t-SNE is widely used in various fields, including machine learning, biology, and natural language processing, for visualizing high-dimensional data, identifying clusters, and uncovering patterns that may not be apparent in the original space. It excels at revealing local structures and is especially valuable for data exploration and preliminary analysis. However, it's important to note that t-SNE does not preserve global distances and may not be suitable for certain tasks like distance-based clustering or regression.

## What is eigendecomposition?

Eigendecomposition is a fundamental concept in linear algebra that involves breaking down a square matrix into its constituent parts, primarily eigenvectors and eigenvalues. Here's a simplified explanation:

When you have a square matrix (a table of numbers), eigendecomposition helps you find special vectors (eigenvectors) and corresponding values (eigenvalues) that, when multiplied together, give you back the original matrix. These eigenvectors represent the directions along which the matrix behaves like simple stretching or compression, while the eigenvalues tell you how much that stretching or compression occurs. In practical terms, eigendecomposition is widely used in various fields, including physics, engineering, and data analysis, to understand and manipulate linear transformations in a more intuitive and computationally efficient way

## What is singular value decomposition?

Singular Value Decomposition (SVD) is a fundamental matrix factorization technique used in linear algebra and data analysis. It decomposes a matrix into three simpler matrices, providing insights into the properties and structure of the original matrix. Here's a simplified explanation:

Suppose you have a rectangular matrix, like a spreadsheet of data. SVD helps you break this matrix down into three matrices: $U$, $Σ$, and $V^T$ (the transpose of V). 

- `U` contains information about how the data points relate to each other

- `Σ` contains information about the "strength" or "importance" of each relationship 

- $V^T$ helps you understand how these relationships relate to the original features. 


Together, these three matrices provide a way to understand and represent the original data more efficiently, reduce noise, and reveal underlying patterns or structures. SVD has numerous applications, including dimensionality reduction, image compression, and collaborative filtering in recommendation systems.


## What is the difference between eig and svd?

**Eigendecomposition (Eig)**: 

Eigendecomposition is a matrix factorization technique mainly applied to square matrices. It breaks a matrix into its eigenvectors (directions) and eigenvalues (scaling factors) to understand how the matrix stretches or compresses data along specific directions. Eigendecomposition is primarily used to analyze linear transformations, making it useful in physics, engineering, and geometric transformations, but it's not suitable for non-square or rectangular matrices.



**Singular Value Decomposition (SVD)**: 

SVD is a matrix factorization technique applicable to both square and rectangular matrices. It decomposes a matrix into three constituent matrices: `U`, `Σ`, and $V^T$. `U` and `V` contain orthogonal vectors, while `Σ` is a diagonal matrix with singular values quantifying the importance of each vector direction. SVD is versatile and widely used in various applications, including data compression, dimensionality reduction, noise reduction, and recommendation systems, making it suitable for a broad range of data analysis tasks and any matrix shape.

## What is a manifold?

A manifold is a fundamental concept in mathematics and geometry, and it has applications in various fields, including data analysis and machine learning. In simple terms, a manifold is a mathematical space that locally resembles Euclidean space (ordinary flat space), even though it may have a more complex global structure.

Imagine you have a curved surface, like the surface of a sphere. When you look closely at a small portion of the sphere, it seems almost flat, like a piece of paper. This "local flatness" is what a manifold represents. It's a way to describe spaces that, when examined up close, behave like familiar flat spaces.

Manifolds can have various shapes and dimensions, and they are used in data analysis to understand and visualize complex data structures. Techniques like t-Distributed Stochastic Neighbor Embedding (t-SNE) and Isomap are used to map high-dimensional data to lower-dimensional manifolds, making it easier to explore and analyze intricate data patterns. Manifolds help us uncover the underlying structure of data in a more manageable way, particularly when the data is not linearly separable or has non-linear relationships.

## Task 0:  PCA


Principal Component Analysis (PCA) is a **dimensionality reduction technique used in machine learning and statistics to simplify a dataset while retaining as much of the original information (variance) as possible**. 

The main idea is to transform the data into a **new coordinate system where the greatest variance lies on the first coordinate** (the first principal component), the second greatest variance on the second coordinate, and so on.

### Key Concepts:

#### Dimensionality Reduction

In datasets with many features, some features might be redundant or irrelevant. PCA reduces the number of dimensions while preserving the essential structure of the data.

#### Variance 
PCA seeks to maximize the variance captured by the new features (principal components). The principal components are linear combinations of the original features, ordered by the amount of variance they capture.

#### Orthogonal Transformation

PCA transforms the original data into a set of orthogonal (uncorrelated) components. This means the new axes (principal components) are perpendicular to each other.

#### Singular Value Decomposition (SVD)

PCA can be computed using SVD, which decomposes the data matrix into three matrices (U, S, and V). This decomposition helps in finding the directions (principal components) that capture the maximum variance.

#### Variance Threshold

Often, we don’t need all the principal components. By specifying a variance threshold (e.g., 95%), PCA can select the minimum number of components required to retain that fraction of the total variance.

### How PCA Works Step-by-Step

#### 1. Standardize the Data 

Mean-centering the data is crucial. PCA assumes the data is centered around the origin, meaning each feature has a mean of 0.

#### 2. Compute the Covariance Matrix 

This matrix expresses how much each feature varies from the mean concerning other features.

#### 3. Perform SVD 

Decompose the covariance matrix using SVD. This gives you matrices (U, S, and V) where:

- `U` contains the left singular vectors (principal components).
- `S` contains singular values (related to the variance explained by each component).
- `V` (or `Vh`) contains the right singular vectors.

#### 4. Determine the Number of Components 

Using the cumulative sum of singular values, determine the smallest number of principal components that retain the desired variance.

#### 5. Construct the Projection Matrix 

Form the matrix that projects the original data into the space spanned by the selected principal components

### How PCA Constructs the Principal Components

As there are as many principal components as there are variables in the data, principal components are constructed in such a manner that the first principal component accounts for the largest possible variance in the data set. 

For example, let's assume that the scatter plot of our data set is as shown below, can we guess the first principal component ? Yes, it's approximately the line that matches the purple marks because it goes through the origin and it’s the line in which the projection of the points (red dots) is the most spread out. 

Or mathematically speaking, it's the line that maximizes the variance (the average of the squared distances from the projected points (red dots) to the origin).
