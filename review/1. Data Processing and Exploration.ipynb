# Data Processing and Exploration

Data processing and exploration are critical steps in machine learning that involve **preparing and understanding the data** before building models. These steps ensure the data is **clean**, **accurate**, and **relevant**, which directly impacts the performance of the machine learning models. 

Proper data processing helps in **handling missing values**, **normalizing data**, and **encoding categorical variables**, while data **exploration allows for identifying patterns**, trends, and potential issues in the dataset. 

<img src="../figs/review/data.png" alt="logo" width="500"/>


- **Data Cleaning**: Removing or correcting inaccuracies and handling missing values to ensure the dataset is accurate and complete.
- **Data Transformation**: Normalizing and encoding data to convert it into a format suitable for analysis and model building.
- **Data Exploration**: Analyzing the data to identify patterns, trends, and potential problems, which informs better model selection and tuning.


### What steps do you take to preprocess and clean data before applying machine learning algorithms?

Preprocessing and cleaning data before applying machine learning algorithms involve several essential steps to ensure the data is suitable for analysis and model building. This process **improves data quality, helps avoid biases, and increase model performance**. 

- **Handling Missing Values**: Filling in or removing missing data to ensure completeness, using methods like mean imputation or deleting rows/columns with missing values.
- **Removing Duplicates and Outliers**: Identifying and eliminating duplicate entries and outliers to maintain data integrity and avoid skewed results.
- **Data Transformation**: Normalizing, scaling, and encoding categorical variables to convert data into formats suitable for machine learning algorithms, ensuring consistent and comparable data inputs.


**We will illustrate each of these key concepts with code snippets from the famous `Titanic` dataset**

#### Handling missing values

```python
import pandas as pd
titanic = pd.read_csv('titanic.csv')

# Display missing values
print(titanic.isnull().sum())

# Fill missing 'Age' values with the mean
titanic['Age'].fillna(titanic['Age'].mean(), inplace=True)

# Drop rows with missing 'Cabin' values (if many are missing)
titanic.drop(columns=['Cabin'], inplace=True)
```
####  Removing Duplicates and Outliers

Removing duplicate entries and identifying outliers is crucial for maintaining data integrity and avoiding skewed results.

```python
# Remove duplicate rows
titanic.drop_duplicates(inplace=True)

# Identify outliers in 'Fare' using the IQR method
Q1 = titanic['Fare'].quantile(0.25)
Q3 = titanic['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Filter out outliers
titanic = titanic[~((titanic['Fare'] < (Q1 - 1.5 * IQR)) | (titanic['Fare'] > (Q3 + 1.5 * IQR)))]
```

### How do you normalize and standartize data

Normalizing and standardizing data are essential preprocessing steps in machine learning that ensure features are on a similar scale, improving model performance and training stability. 

**Normalization typically scales the data to a range of `[0, 1]`**, making it useful for algorithms sensitive to feature scales, such as k-nearest neighbors and neural networks. 

**Standardization, on the other hand, transforms data to have a mean of 0 and a standard deviation of 1**, which is particularly important for algorithms that assume normally distributed data, like linear regression and principal component analysis. 

These transformations help models converge faster and make the optimization process more efficient.

- **Normalization**: Scaling data to a fixed range, usually `[0, 1]`, using min-max scaling.
- **Standardization**: Transforming data to have a mean of `0` and a standard deviation of `1`.
- **Improvement in Model Performance**: Ensures that all features contribute equally to the result, improving the effectiveness of gradient descent and other optimization algorithms.

#### Normalization

```python
from sklearn.preprocessing import MinMaxScaler

# Example data
data = titanic[['Age', 'Fare']]

# Normalize the data to [0, 1] range
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(data)

# Convert back to DataFrame for readability
normalized_df = pd.DataFrame(normalized_data, columns=['Age', 'Fare'])
print(normalized_df.head())
```

#### Standartization

```python
from sklearn.preprocessing import StandardScaler

# Example data
data = titanic[['Age', 'Fare']]

# Standardize the data to mean 0 and standard deviation 1
scaler = StandardScaler()
standardized_data = scaler.fit_transform(data)

# Convert back to DataFrame for readability
standardized_df = pd.DataFrame(standardized_data, columns=['Age', 'Fare'])
print(standardized_df.head())
```

### What are some common techniques for data visualization?

Effective visualization techniques can reveal insights that may not be immediately apparent through raw data analysis. Common techniques include bar charts for categorical data comparison, histograms for distribution analysis, scatter plots for identifying relationships between variables, and box plots for summarizing the distribution and identifying outliers. These visual tools help in making informed decisions about data preprocessing, feature selection, and model development.

- **Bar Charts**: Used for comparing the frequency or count of categorical data.
- **Histograms**: Display the distribution of a continuous variable, showing the frequency of data points within specified ranges.
- **Scatter Plots**: Show the relationship between two continuous variables, helping to identify correlations and patterns.
- **Box Plots**: Summarize the distribution of a dataset and highlight the median, quartiles, and potential outliers.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart of survival counts
sns.countplot(data=titanic, x='Survived')
plt.title('Survival Counts')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

### What is feature engineering, and why is it important?

Feature engineering is the **process of using domain knowledge to create new features or modify existing ones** to improve the performance of machine learning models. 

It involves transforming raw data into meaningful features that better represent the underlying problem to the predictive models. This step is crucial because the quality and relevance of features directly impact the effectiveness of the model.

- **Creation of New Features**: Deriving new variables from the existing data that can reveal hidden patterns and improve model accuracy.
- **Transformation of Features**: Modifying features through normalization, scaling, or encoding to ensure they are in a format suitable for the algorithm.
- **Feature Selection**: Identifying and retaining the most relevant features to reduce dimensionality and improve model performance.

#### Creating New Features

```python
# Creating a new feature 'FamilySize' as the sum of 'SibSp' and 'Parch'
titanic['FamilySize'] = titanic['SibSp'] + titanic['Parch'] + 1
print(titanic[['SibSp', 'Parch', 'FamilySize']].head())
```

#### Transformation of Features

```python
# Log transformation of 'Fare' to reduce skewness
import numpy as np

titanic['Fare_log'] = np.log1p(titanic['Fare'])
print(titanic[['Fare', 'Fare_log']].head())
```

#### Encoding Categorical Variables
```python
# Encoding 'Sex' as binary feature
titanic['Sex_encoded'] = titanic['Sex'].map({'male': 0, 'female': 1})
print(titanic[['Sex', 'Sex_encoded']].head())

```
