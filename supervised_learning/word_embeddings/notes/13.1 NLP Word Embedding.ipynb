# Natural Language Processing - Word Embeddings

## What is natural language processing?

Natural Language Processing (**NLP**) is a field of artificial intelligence that **focuses on the interaction between computers and human language**. It involves the development of algorithms and models to enable machines to understand, interpret, and generate human language in a way that is both meaningful and contextually relevant.

Applications of NLP include:

1. **Machine Translation**: Translating text from one language to another, like Google Translate.


2. **Sentiment Analysis**: Determining the emotional tone of a piece of text, often used for social media monitoring and customer feedback analysis.


3. **Speech Recognition**: Converting spoken language into written text, used in virtual assistants like Siri and Alexa.


4. **Named Entity Recognition (NER)**: Identifying and classifying entities such as names of people, places, and organizations in text.


5. **Text Summarization**, **Text Classification** and **Text Generation**: 
- Generating concise summaries of long documents, helpful for news articles and research papers.
- Categorizing text into predefined classes, used for spam filtering and content categorization.
- Creating human-like text, used for various applications, including creative writing and content generation.

6. **Chatbots and Virtual Assistants**: Creating conversational agents that can interact with users in natural language.

NLP has extensive applications across various industries, including healthcare, finance, customer service, education, and more. It continues to advance rapidly, improving the ways in which computers interact with and understand human language.


## What is a word embedding?

A **word embedding is a dense, low-dimensional vector representation of a word in natural language**.


It is designed to capture semantic and contextual information, enabling computers to understand relationships between words and their meanings.


**Example**:
Consider the words "`king`," "`queen`," "`man`," and "`woman`." Word embeddings can represent these words as vectors in a multi-dimensional space. Words with similar meanings or contextual usage will have closer vector representations.

Illustration with Python code using Word2Vec (a popular word embedding technique):


```python
# Install gensim library if you haven't already: pip install gensim

from gensim.models import Word2Vec

# Sample sentences for training the Word2Vec model
sentences = [
    ["king", "queen", "royal"],
    ["man", "woman", "person"],
    ["apple", "orange", "fruit"],
    ["car", "bike", "vehicle"]
]

# Training the Word2Vec model on the given sentences
model = Word2Vec(sentences, vector_size=3, window=2, min_count=1, sg=1)

# Getting the word embedding vector for a specific word
word_embedding = model.wv["king"]
print(word_embedding)
```

## What is bag of words?

Bag of Words (BoW) is a common technique in natural language processing for text representation. **It converts text documents into a numerical format by creating a "bag" or collection of words present in the document, ignoring grammar and word order**. 

Each word in the document is represented as a unique feature, and the frequency of each word is used as its value.

Example:

Consider the following two simple sentences:

- `I love natural language processing.`
- `Natural language processing is fascinating.`

BoW representation for these sentences:

| Sentence | I | love | natural | language | processing | is | fascinating |
|----------|---|------|---------|---------|------------|----|-------------|
| 1        | 1 | 1    | 1       | 1       | 1          | 0  | 0           |
| 2        | 0 | 0    | 1       | 1       | 1          | 1  | 1           |

*In the BoW representation, each column represents a unique word present in the sentences, and each row corresponds to a sentence. The numbers in the table indicate the frequency of each word in the respective sentence. The order of words in the original sentences is disregarded, and only the presence and count of words matter*.





## What is TF-IDF?

TF-IDF stands for Term Frequency-Inverse Document Frequency. It is a **numerical representation** technique used in natural language processing **to evaluate the importance of a word in a document relative to a collection of documents (corpus)**. TF-IDF combines two factors:

- **Term Frequency (TF)**: The number of times a word appears in a document, indicating its importance within that specific document.


- **Inverse Document Frequency (IDF)**: The logarithmically scaled inverse fraction of documents that contain the word, determining the general importance of the word across the entire corpus.


The TF-IDF score is higher for words that are frequent in the specific document but relatively rare across the entire corpus, highlighting words that are both relevant to the document and distinguishable from other documents in the collection.

Example:

Consider:

- Document 1: `I love natural language processing.`
- Document 2: `Natural language processing is fascinating.`

TF-IDF representation:

|        | I       | love    | natural | language | processing | is      | fascinating |
|--------|---------|---------|---------|----------|------------|---------|-------------|
| Doc 1  | 0.301   | 0.301   | 0.301   | 0.301    | 0.301      | 0.0     | 0.0         |
| Doc 2  | 0.0     | 0.0     | 0.301   | 0.301    | 0.301      | 0.426   | 0.426       |

## What is CBOW?

CBOW stands for Continuous Bag of Words. It is a type of word embedding technique used in natural language processing. 

**CBOW aims to predict a target word from its surrounding context words**. It takes a context window of neighboring words as input and tries to learn the representation of the target word based on the context.

Example:

- Consider the sentence: `The cat sits on the mat.`

For a context window size of 2 (2 words to the left and 2 words to the right), CBOW will create training data and target pairs like this:

| Context (Input)       | Target (Output) |
|-----------------------|-----------------|
| The cat               | sits            |
| cat sits on           | the             |
| sits on the mat       | the             |


CBOW will use this data to learn word embeddings that can predict the target word ("sits" or "the") from the given context.

*CBOW is especially useful for generating word representations for small datasets and is computationally efficient* compared to other word embedding techniques like Skip-gram, which tries to predict context words from a given target word.


## What is a skip-gram?

Skip-gram is another type of word embedding technique used in natural language processing. Unlike CBOW, which predicts a target word from its surrounding context words, Skip-gram does the opposite. **It takes a target word as input and aims to predict the context words in its vicinity**.

Example:

- Consider the sentence: `The cat sits on the mat.`

For a context window size of 2 (2 words to the left and 2 words to the right), Skip-gram will create training data and target pairs like this:

| Target (Input) | Context (Output)          |
|----------------|---------------------------|
| sits           | The, cat, on              |
| the            | cat, sits, on, mat        |


Skip-gram will use this data to learn word embeddings that can predict the context words ("The, cat, on" or "cat, sits, on, mat") from a given target word.

*Skip-gram is often used in scenarios where we want to generate better word representations for infrequent words or in cases where the focus is on understanding the context rather than individual words*.

## What is an n-gram?

An n-gram is a **contiguous sequence of n items (characters, words, or tokens) from a text or speech**. 

In natural language processing, n-grams are used to **analyze the co-occurrence of words or characters in a given piece of text**. Commonly, n-grams are used to capture the context and syntax of sentences.

Example:
- `The quick brown fox jumps over the lazy dog.`

Examples of n-grams with different values of n:

- 1-gram (unigram): `["The", "quick", "brown", "fox", "jumps", "over", "the", "lazy", "dog"]`


- 2-gram (bigram): [`"The quick"`, `"quick brown"`, `"brown fox"`, `"fox jumps"`, `"jumps over"`, `"over the"`, `"the lazy"`, `"lazy dog"`]


- 3-gram (trigram): ["`The quick brown`", "`quick brown fox`", "`brown fox jumps`", "`fox jumps over`", "`jumps over the`", "`over the lazy`", "`the lazy dog`"]


In this example, the unigrams represent individual words, bigrams represent two-word sequences, and trigrams represent three-word sequences from the given sentence. N-grams provide a way to understand the local relationships and patterns within the text, which can be helpful in various NLP tasks.






N-grams are widely used in various NLP tasks, including language modeling, text classification, and information retrieval. They help in capturing local word order and can provide valuable insights into the structure and patterns within a text.

## What is negative sampling?

Negative sampling is a **technique used in training word embeddings, especially in skip-gram based models, to efficiently handle large vocabularies**. 

Instead of considering all words as potential context words during training, **negative sampling randomly selects a small number of negative samples (non-context words) for each target word**, making the training process more computationally efficient.

Example:

Let's assume we have a skip-gram model with a target word "`apple`" and a vocabulary of 10,000 words. Instead of using all 10,000 words as context words during training, negative sampling randomly selects a small subset of words (e.g., 5-20 words) as negative samples for the target word "apple." These negative samples are used to contrast with the positive samples (actual context words) during training.

By reducing the number of context words and using negative samples, the training process becomes more efficient without compromising the quality of the word embeddings. Negative sampling is a critical component in popular word embedding models like Word2Vec, where it allows for effective learning of high-quality word representations.


## What is word2vec, GloVe, fastText, ELMo?

Word2Vec, GloVe, fastText, and ELMo are all popular word embedding techniques in natural language processing (NLP).


- `Word2Vec` is based on two models: Continuous Bag of Words (CBOW) and Skip-gram, used to learn word vectors from large text corpora.


- `GloVe` (Global Vectors for Word Representation) uses co-occurrence statistics to create word embeddings that capture global word relationships.


- `fastText` extends Word2Vec by representing words as bags of character n-grams, enabling better handling of out-of-vocabulary words and subword information.


- `ELMo` (Embeddings from Language Models) is a contextual word embedding model that generates word representations based on the surrounding context, providing context-aware embeddings for words.

# Performing Sentiment Analysis for IMDB Movie Review

## IMBD Dataset

IMDb (Internet Movie Database) is an **online platform that provides a comprehensive database of movies, TV shows, and other entertainment content**.

IMDb movie reviews refer to **user-generated reviews and ratings** for movies available on the platform. These reviews allow users to express their opinions and share their experiences regarding the quality, plot, acting, and other aspects of movies.

IMDb movie reviews are typically **short texts written by viewers**, and they reflect a range of sentiments, including positive and negative feedback.

With a large user base and a vast collection of movies, IMDb movie reviews offer valuable insights into the public perception and reception of films, serving as a valuable resource for understanding audience preferences and opinions.

```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Flatten, Dense
from tensorflow.keras.optimizers import Adam
```

## Load the IMDb dataset


```python
max_vocab_size = 20000
(x_train, y_train), (x_test, y_test) = imdb.load_data(num_words=max_vocab_size)
```

## Explore the dataset

### How the data are stored

**Movie reviews are converted into numbers** using a process called text encoding. In this process, **each word in a movie review is assigned a unique numerical index** based on its position in a predefined vocabulary. 

The vocabulary consists of a set of frequently occurring words in the dataset. The **movie review is then represented as a sequence of these numerical indices**, where each index corresponds to a specific word in the review. 

This numerical representation allows the movie reviews to be stored as sequences of numbers, which can be efficiently processed and used as input to machine learning models, such as recurrent neural networks (RNNs), for sentiment analysis and other natural language processing tasks.


### Display random reviews

Exploring a **subset of randomly selected reviews from the IMDB dataset** offers a glimpse into the diversity of opinions and sentiments expressed by viewers. 

These reviews represent a mix of positive and negative feedback, providing valuable insights into the subjective experiences and perceptions of different movies. By analyzing the displayed reviews, we can observe the varying lengths, writing styles, and overall sentiment conveyed by users

```python
import numpy as np

# Reverse word index mapping
word_index = imdb.get_word_index()
reverse_word_index = {index: word for word, index in word_index.items()}

# Get 5 random review indices
random_indices = np.random.choice(len(x_test), size=5, replace=False)

# Retrieve and print the corresponding review text
for index in random_indices:
    review = [reverse_word_index.get(i - 3, '') for i in x_test[index]]
    review_text = ' '.join(review)
    print(f'Review {index + 1}:')
    print(review_text)
    print('\n' + '-' * 50 + '\n')
```

### Display the distribution of movie review sentiment

```python
# Concatenate the labels of the training and testing sets
labels = np.concatenate((y_train, y_test), axis=0)

# Count the occurrences of each class
class_counts = np.bincount(labels)

# Define class labels
class_labels = ['Negative', 'Positive']

# Plot the bar chart
plt.bar(class_labels, class_counts)
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.title('Class Distribution of IMDB Movie Reviews')
```

## Text Preprocessing

We will use the `pad_sequences` function to ensure that all sequences in our training and test data have a fixed length of `200` tokens. 

This is important because neural networks require inputs of the same size to process data efficiently. 

Shorter sequences are padded with zeros, and longer sequences are truncated to meet this uniform length. 

By standardizing the data in this way, we can effectively train our sentiment analysis model on the IMDb movie review dataset and achieve consistent and reliable results.

```python
max_sequence_length = 200
x_train = pad_sequences(x_train, maxlen=max_sequence_length)
x_test = pad_sequences(x_test, maxlen=max_sequence_length)
```

## Word Embedding

The embedding layer transforms words into continuous, dense vectors, capturing their semantic meaning and contextual relationships. 

By mapping words to numerical representations, the model gains the ability to recognize similarities between words based on their usage, enhancing performance in NLP tasks like sentiment analysis. 

These learned embeddings provide valuable transferable knowledge, making the model more adaptable and effective in understanding text data.

```python
embedding_dim = 100

model = Sequential()
model.add(Embedding
          (
              input_dim=max_vocab_size, 
              output_dim=embedding_dim, 
              input_length=max_sequence_length
          )
         )

model.add(Flatten())
```

## Model Building

Next, we construct the final layers of the neural network model for sentiment analysis. 

- The first Dense layer with `128` units and the ReLU activation function captures complex patterns in the data


- The second Dense layer with `1` unit and the sigmoid activation function outputs a probability score, indicating the sentiment polarity (positive or negative) of the movie review text.

```python
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

## Model Training

Now we prepare the neural network model for sentiment analysis. 

We set the binary cross-entropy loss function, suitable for binary classification tasks like sentiment analysis, and employ the Adam optimizer with a learning rate of 0.0001 for weight updates during training. 


Additionally, we select `accuracy` as the evaluation metric to measure the model's performance.

```python
model.compile(
    loss='binary_crossentropy', 
    optimizer=Adam(learning_rate=0.0001), 
    metrics=['accuracy'])
model.summary()
```

Finally, we train the neural network model for sentiment analysis on the provided training data (`x_train` and `y_train`). 


We set the batch size to 64, meaning that during training, the model will update its weights after processing 64 samples at a time. The epochs parameter is set to 5, indicating that the entire training dataset will be processed five times during training. 


The `model.fit()` function initiates the training process, optimizing the model's parameters based on the specified loss function and optimizer. 


We also use the validation data (`x_test` and `y_test`) to evaluate the model's performance after each epoch, ensuring that it generalizes well to unseen data.

```python
batch_size = 64
epochs = 5
model.fit(
    x_train, 
    y_train, 
    batch_size=batch_size, 
    epochs=epochs, 
    validation_data=(x_test, y_test)
)
```

## Model Evaluation

To conclude, we evaluate the trained neural network model's performance on the test dataset and print the test accuracy, representing how well the model generalizes to unseen data and performs sentiment analysis on the IMDb movie review dataset.

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f"Test accuracy: {accuracy}")
```
