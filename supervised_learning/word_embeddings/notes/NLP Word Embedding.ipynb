## Why Do We Need Word Embeddings?

Computers don‚Äôt understand words like *algorithm*, *data*, or *Python* the way we do. To process text, we need to **convert words into numbers** so machine learning models can work with them.

### From Text to Numbers

A simple idea is to assign each word a unique number:

| Word     | ID |
| -------- | -- |
| neural   | 1  |
| network  | 2  |
| machine  | 3  |
| learning | 4  |

But this doesn‚Äôt tell the model that *neural* and *network* are related, or that *Python* and *programming* often appear together.

### Why Embeddings?

Word embeddings solve this by representing words as **vectors** (lists of numbers) that capture **meaning** and **relationships**.

- Words with similar meanings (like *training* and *learning*) are close together.
- Unrelated words (like *loop* and *database*) are far apart.

This helps algorithms understand context and make smarter predictions.

### How Do We Create Word Embeddings?

There are different techniques:

1. **Bag of Words (BoW)** ‚Äì Counts how often words appear in a document.
2. **TF-IDF** ‚Äì Adjusts for word importance across documents.
3. **Word2Vec** ‚Äì Learns dense vectors where similar words are grouped together.
4. **GloVe** ‚Äì Builds embeddings using global co-occurrence of words.
5. **FastText** ‚Äì Handles rare words by breaking them into smaller parts.

Each step moves us closer to **teaching machines to understand language** instead of just memorizing it.

## What is Bag of Words?

Bag of Words (BoW) is a simple and widely used technique in natural language processing to represent text. **It transforms text into a numerical format by creating a "bag" of all unique words, ignoring grammar and word order.**

Each word becomes a feature, and its value is the number of times it appears in the document.

**Example:**

Two sentences:

* `I love natural language processing.`
* `Natural language processing is fascinating.`

| Sentence | I | love | natural | language | processing | is | fascinating |
| -------- | - | ---- | ------- | -------- | ---------- | -- | ----------- |
| 1        | 1 | 1    | 1       | 1        | 1          | 0  | 0           |
| 2        | 0 | 0    | 1       | 1        | 1          | 1  | 1           |

> *Here, each column is a unique word. Each row shows how many times the word appears in the sentence. Word order and grammar are ignored.*

### Pros and Cons

- **Simple and effective** for small datasets.
- **Doesn‚Äôt capture meaning or context** (e.g., *‚Äúgood‚Äù* and *‚Äúgreat‚Äù* are unrelated).

---


## What is TF-IDF?

**TF-IDF** (Term Frequency‚ÄìInverse Document Frequency) is a technique to measure how important a word is in a document **relative to a collection of documents (corpus)**.

It combines two ideas:

* **Term Frequency (TF):** How often a word appears in a document.
* **Inverse Document Frequency (IDF):** How rare the word is across all documents.

The formula gives higher scores to words that are frequent in a document but uncommon in the corpus. This helps highlight **important and distinctive words**.

**Example:**

Documents:

* Doc 1: `I love natural language processing.`
* Doc 2: `Natural language processing is fascinating.`

|       | I    | love | natural | language | processing | is   | fascinating |
| ----- | ---- | ---- | ------- | -------- | ---------- | ---- | ----------- |
| Doc 1 | 0.30 | 0.30 | 0.30    | 0.30     | 0.30       | 0.00 | 0.00        |
| Doc 2 | 0.00 | 0.00 | 0.30    | 0.30     | 0.30       | 0.42 | 0.42        |

> *Common words like ‚Äúis‚Äù get low scores, while rare but relevant words like ‚Äúfascinating‚Äù score higher.*

### Why use TF-IDF?

- Highlights **distinctive words** in each document.
- Reduces the influence of **common words** (e.g., *the*, *is*, *and*).
- Still doesn‚Äôt capture **word meaning or context** like embeddings do.

## Why Do We Need Word Vectors?

BoW and TF-IDF represent text as high-dimensional vectors but don‚Äôt capture **meaning** or **relationships between words**. For example, *‚Äúalgorithm‚Äù* and *‚Äúmodel‚Äù* are treated as unrelated even though they‚Äôre semantically similar.

**Word vectors** solve this by representing words as dense, low-dimensional vectors where **similar words are closer together in space**.

### Benefits of Word Vectors:

-  **Capture meaning and relationships** (e.g., *‚Äúdata‚Äù* and *‚Äúinformation‚Äù* are close).
-  **Efficient and compact** compared to sparse representations.
-  Enable advanced NLP tasks like analogies (*‚ÄúPython‚Äù - ‚Äúprogramming‚Äù + ‚Äústatistics‚Äù ‚âà ‚ÄúR‚Äù*).

## What is CBOW?

**CBOW** (Continuous Bag of Words) is a word embedding technique that learns word representations by predicting a target word from its surrounding context.

It uses a **window of neighboring words** as input and tries to guess the missing word. This helps the model learn relationships between words based on their usage in text.

**Example:**

Consider the sentence:
`Deep learning models achieve great results in NLP.`

For a context window size of 2 (2 words to the left and right), CBOW creates training pairs like this:

| Context (Input)              | Target (Output) |
| ---------------------------- | --------------- |
| Deep learning                | models          |
| learning models achieve      | great           |
| models achieve great results | in              |

The model learns to predict the target word (*‚Äúmodels‚Äù*, *‚Äúgreat‚Äù*, *‚Äúin‚Äù*) from its context.


**Why CBOW?**

* Efficient for large corpora.
* Learns meaningful word embeddings even from small datasets.
* Useful in machine learning pipelines to map words to vectors.

## What is Skip-gram?

**Skip-gram** is a word embedding technique that does the opposite of CBOW: it predicts **context words** given a **target word**.

It learns to generate word representations by trying to guess the surrounding words of a given word in a sentence.


**Example:**

Consider the sentence:
`Neural networks power many AI applications.`

For a context window size of 2 (2 words to the left and right), Skip-gram creates training pairs like this:

| Target (Input) | Context (Output) |
| -------------- | ---------------- |
| neural         | networks         |
| neural         | power            |
| networks       | neural           |
| networks       | power            |
| power          | networks         |
| power          | many             |

Here, the model tries to predict the surrounding context words (*‚Äúnetworks‚Äù*, *‚Äúpower‚Äù*, *‚Äúmany‚Äù*) for each target word (*‚Äúneural‚Äù*, *‚Äúpower‚Äù*).

**Why Skip-gram?**

* Performs well with **rare words**.
* Useful when the dataset is large and diverse.
* Creates embeddings that capture relationships between words effectively.

## What is Word2Vec?

**Word2Vec** is a powerful technique in natural language processing that learns dense vector representations of words based on their context in a large corpus. These vectors capture semantic relationships, so words with similar meanings are placed closer together in the vector space.

Word2Vec uses two main architectures to learn embeddings:

* **Continuous Bag of Words (CBOW):** Predicts a target word from its surrounding context words.
* **Skip-gram:** Predicts surrounding context words given a target word.

For example, in a machine learning corpus, given the sentence:
`Neural networks are widely used in deep learning.`

* **CBOW** could use the context (*‚Äúneural‚Äù, ‚Äúare‚Äù, ‚Äúwidely‚Äù, ‚Äúin‚Äù*) to predict the word *‚Äúnetworks‚Äù*.
* **Skip-gram** could use the target word *‚Äúnetworks‚Äù* to predict nearby words like *‚Äúneural‚Äù* and *‚Äúdeep‚Äù*.

These embeddings are widely used in ML pipelines for tasks like text classification, sentiment analysis, and machine translation.


## What is FastText?

**FastText** is a word embedding technique that represents words as a collection of smaller units called **subwords** or **character n-grams**. Instead of learning a single vector for each whole word, FastText learns vectors for these subwords and combines them to form the word‚Äôs overall embedding.

This method allows FastText to:

* Handle **rare or unseen words** by building their embeddings from known subwords.
* Capture **morphological variations** like prefixes and suffixes (e.g., *run*, *running*, *runner*).
* Be more robust to **misspellings** or slight variations in words.

### Example:

Consider the word **‚Äúembedding‚Äù** using character 3-grams (trigrams) with boundary markers `<` and `>`:

`<em`, `emb`, `mbe`, `bed`, `edd`, `ddi`, `din`, `ing`, `ng>`

FastText learns vectors for each trigram and sums them to create the vector for *‚Äúembedding‚Äù*.

So even if *‚Äúembedding‚Äù* doesn‚Äôt appear in the training data, FastText can still generate a meaningful vector because it understands the parts that make up the word.

This subword approach makes FastText especially useful for domains with lots of rare or technical terms, such as machine learning or programming languages.


## Summary of Text Representation Techniques

| Technique              | Representation Type            | Key Idea                                   | Strengths                       | Limitations                                |
| ---------------------- | ------------------------------ | ------------------------------------------ | ------------------------------- | ------------------------------------------ |
| **Bag of Words (BoW)** | Sparse count vectors           | Counts how often each word appears         | Simple, easy to implement       | Ignores word order and meaning             |
| **TF-IDF**             | Sparse weighted vectors        | Weighs word frequency by rarity            | Highlights important words      | Still ignores word meaning/context         |
| **Word2Vec**           | Dense word embeddings          | Learns word meaning from local context     | Captures semantic relationships | Can‚Äôt handle rare or misspelled words well |
| **FastText**           | Dense embeddings with subwords | Builds word vectors from character n-grams | Handles rare/misspelled words   | More complex, slightly slower              |


## Overview

- **Bag of Words (BoW)** is the simplest method that represents text by counting word occurrences. It‚Äôs easy to implement but treats words independently and ignores grammar and word order.

- **TF-IDF** improves on BoW by weighting words according to their importance in a document relative to the whole corpus. It highlights distinctive words but still doesn‚Äôt capture word meaning.

- **Word2Vec** creates dense vector representations of words by learning from their local context within sentences. This lets it capture semantic relationships, so similar words have similar vectors. However, it treats each word as an atomic unit, struggling with rare or unseen words.

- **FastText** extends Word2Vec by representing words as a combination of character-level subwords (n-grams). This enables it to build meaningful embeddings even for rare, misspelled, or new words by composing their vectors from known subparts.

## Why Do We Need Contextual Word Embeddings?

Traditional word embeddings like Word2Vec and FastText assign each word a **single fixed vector**, regardless of where or how the word appears. This means the word *‚Äúcode‚Äù* will have the same representation whether it refers to **programming code** or a **discount code**.

However, in human language, many words have **multiple meanings** depending on their context. To truly understand language, machines need embeddings that can **change dynamically based on surrounding words**.

### The Challenge

* Words can be **ambiguous** (e.g., *‚Äúbank‚Äù* of a river vs. *‚Äúbank‚Äù* as a financial institution).
* The **meaning of a word depends on its sentence or document context**.
* Static embeddings **fail to capture this nuance**, limiting performance on many NLP tasks.

### The Solution: Contextual Embeddings

Contextual embeddings represent each word **as a vector influenced by its specific context** in a sentence or paragraph. Instead of a single fixed vector, the embedding for a word like *‚Äúcode‚Äù* adapts based on whether the sentence talks about software or discounts.

This approach enables models to better understand meaning, disambiguate words, and improve performance across a wide range of language tasks.

## How Do Transformers Enable Contextual Embeddings?

Transformers are a modern neural network architecture designed to better understand language by analyzing entire sentences at once. Unlike earlier models that read text word-by-word in sequence, transformers look at all words simultaneously and learn how each word relates to every other word in the sentence. This ability comes from a mechanism called **self-attention**, which allows the model to focus on relevant words when building the meaning of each word in context.

For example, consider the sentence:
> *‚ÄúTraining deep neural networks requires large datasets.‚Äù*

When processing the word *‚Äúnetworks‚Äù*, a transformer will pay attention to words like *‚Äúdeep‚Äù* and *‚Äútraining‚Äù* because they provide important clues about the meaning of *‚Äúnetworks‚Äù* here. Similarly, for *‚Äútraining‚Äù*, it considers *‚Äúdatasets‚Äù* to understand the action more precisely.

Because of this, transformers create **contextual embeddings** ‚Äî the vector representation of a word changes depending on the other words around it. This is crucial for handling words with multiple meanings or subtle nuances.

Transformers also use **parallel processing**, which means they analyze all words in a sentence at once, making training much faster than older models like RNNs or LSTMs. By stacking multiple layers of self-attention and processing, transformers gain a deep, layered understanding of complex language patterns.

The result is a powerful way to represent words that captures not just their dictionary meaning but how that meaning shifts based on context, enabling machines to perform much better on tasks like translation, summarization, and question answering.

### üìù Loading and Exploring the Dataset

The first step is to load the collection of markdown files from the folder `ZendeskArticles`. These files contain text data about Holberton School, which we will use as our practical dataset.

This code reads all `.md` files in the folder, stores their contents in a list, and keeps track of filenames for reference.

By printing the total number of documents, the first few filenames, and a snippet of the first document, we get an initial overview of the dataset. This helps us understand the scope and content before applying NLP techniques.

```python
import os

folder_path = "ZendeskArticles"

documents = []
filenames = []

for file in os.listdir(folder_path):
    if file.endswith(".md"):
        file_path = os.path.join(folder_path, file)
        with open(file_path, "r", encoding="utf-8") as f:
            text = f.read()
            documents.append(text)
            filenames.append(file)

print(f"First 3 filenames: {filenames[:3]}")
print("\nFirst 500 characters of the first document:\n")
print(documents[0][:500])
```

### üßê Compute  Basic Statistics

It‚Äôs important to get a quantitative overview of our dataset. This helps us understand its size, diversity, and complexity. In this step, we will calculate:

* **Total documents:** How many text files we have in the collection.
* **Average document length:** The mean number of words per document, giving us a sense of typical document size.
* **Vocabulary size:** The total number of unique words found across all documents, which indicates the range of language used and helps anticipate the dimensionality of our models.

```python
import re
from collections import Counter

# Simple text preprocessing: lowercase + remove punctuation
def clean_text(text):
    text = text.lower()
    text = re.sub(r"[^\w\s]", "", text)  # remove punctuation
    return text

all_words = []
doc_lengths = []

for doc in documents:
    cleaned = clean_text(doc)
    tokens = cleaned.split()
    doc_lengths.append(len(tokens))
    all_words.extend(tokens)

# Stats
total_docs = len(documents)
avg_length = sum(doc_lengths) / total_docs
vocab = set(all_words)
vocab_size = len(vocab)

print(f"Total Documents: {total_docs}")
print(f"Average Document Length: {avg_length:.2f} words")
print(f"Vocabulary Size: {vocab_size} unique words")
print(f"Shortest Document: {min(doc_lengths)} words")
print(f"Longest Document: {max(doc_lengths)} words")
```

### üå•Ô∏è Generate and Display Word Cloud
We createes a word cloud visualization from all documents combined, highlighting the most frequent words with five or more characters.

The process includes cleaning the text by removing markdown links and punctuation, converting to lowercase, filtering out short words, and then generating the word claset.

## Next Step: TF-IDF (Term Frequency-Inverse Document Frequency)

**Bag of Words treats all words equally**, which can cause very common words‚Äîlike ‚Äúsupport‚Äù or ‚Äúarticle‚Äù to dominate the representation, even though they add little meaningful distinction between documents.

TF-IDF addresses this by combining two factors: **it increases the weight of words that appear frequently within a document, while reducing the weight of words that occur across many documents**. This way, TF-IDF highlights terms that are more relevant and discriminative for each individual document.
nt.


```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words="english")

# Fit and transform
X_tfidf = tfidf_vectorizer.fit_transform(documents)

# Get vocabulary
vocab = tfidf_vectorizer.get_feature_names_out()
print(f"üóÇÔ∏è Vocabulary Size: {len(vocab)} words")
```

### üîç Exploring TF-IDF Scores for a Single Document

TF-IDF assigns a score to each word in a document that reflects its importance relative to the entire corpus.

Here, we select one document and extract its TF-IDF vector, then organize the non-zero scores into a ttion.

Sorting the table by TF-IDF values highlights the top 20 words that are most significant in that document, helping us understand which terms best characterize its cotent.


```python
import pandas as pd

# Pick a document (e.g., first one)
doc_idx = 3
tfidf_vector = X_tfidf[doc_idx].toarray()

# Create DataFrame for visualization
df_tfidf = pd.DataFrame(tfidf_vector.T, index=vocab, columns=[f"Doc {doc_idx+1}"])
df_tfidf = df_tfidf[df_tfidf[f"Doc {doc_idx+1}"] > 0]  # Only non-zero scores

# Sort by TF-IDF score
df_tfidf_sorted = df_tfidf.sort_values(by=f"Doc {doc_idx+1}", ascending=False)

print(df_tfidf_sorted.head(20))  # Top 20 words
```

### üìä Comparing Bag of Words and TF-IDF for a Document

This example compares how Bag of Words (BoW) and TF-IDF represent the same document.

We start by displaying the first 300 characters of the chosen document to provide context.

Then, BoW counts and TF-IDF scores are computed separately using their respective vectorizers with English stopwords removed.

The results are combined into a table showing each word‚Äôs frequency (BoW Count) alongside its importance score (TF-IDF Score).

Filtering and sorting highlight the top 15 words, illustrating how TF-IDF adjusts the raw counts to emphasize more informative terms.


```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import pandas as pd


# For demo, pick one document (e.g., index 0)
doc_idx = 0
print(f"üìÑ Document {doc_idx+1} (first 300 chars):\n")
print(documents[doc_idx][:300])

# Create Bag of Words representation
bow_vectorizer = CountVectorizer(stop_words="english")
X_bow = bow_vectorizer.fit_transform(documents)
vocab_bow = bow_vectorizer.get_feature_names_out()

# Create TF-IDF representation
tfidf_vectorizer = TfidfVectorizer(stop_words="english")
X_tfidf = tfidf_vectorizer.fit_transform(documents)
vocab_tfidf = tfidf_vectorizer.get_feature_names_out()

# Get BoW counts for the chosen document
bow_counts = X_bow[doc_idx].toarray().flatten()

# Get TF-IDF scores for the same document
tfidf_scores = X_tfidf[doc_idx].toarray().flatten()

# Combine into a DataFrame for easy comparison
df_compare = pd.DataFrame({
    "Word": vocab_bow,
    "BoW Count": bow_counts,
    "TF-IDF Score": tfidf_scores
})

# Filter out words not in the document (BoW Count = 0)
df_compare = df_compare[df_compare["BoW Count"] > 0]

# Sort by BoW count descending
df_compare_sorted = df_compare.sort_values(by="BoW Count", ascending=False)

# Display top 10 words
print("\n Top 10 words in Document (BoW vs TF-IDF):\n")
print(df_compare_sorted.head(10))
```

## Word2Vec Overview

**Word2Vec** is a technique that transforms words into dense, continuous vectors that capture their semantic meaning based on the context in which they appear. Unlike frequency-based methods like Bag of Words or TF-IDF, Word2Vec learns relationships between words by analyzing how they co-occur with others.

* Two main architectures:

  * **CBOW (Continuous Bag of Words):** Predicts a target word based on its surrounding context words.
  * **Skip-Gram:** Predicts surrounding context words given a target word.

* Words with similar meanings are mapped to nearby vectors, enabling meaningful operations like:
  `king - man + woman ‚âà queen`

* These embeddings encode semantic and syntactic relationships, making them powerful for various NLP tasks such as classification, clustering, and analogy reasoning.


```python
import os
import re
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')


# Simple preprocessing and tokenization
def preprocess(text):
    text = text.lower()
    text = re.sub(r"[^\w\s]", "", text)  # remove punctuation
    tokens = word_tokenize(text)
    return tokens

tokenized_docs = [preprocess(doc) for doc in documents]

# Train Word2Vec model (Skip-gram)
model = Word2Vec(sentences=tokenized_docs, vector_size=100, window=5, min_count=2, sg=1, workers=4)

# Explore embeddings
print("Vocabulary size:", len(model.wv))

# Find most similar words to 'support'
print("\nMost similar to 'holberton':")
print(model.wv.most_similar("holberton", topn=5))
```

### üîç Visualizing Word2Vec Embeddings with t-SNE

To better understand the relationships learned by Word2Vec, we reduce the high-dimensional word vectors to 2D using t-SNE, a dimensionality reduction technique that preserves local structure.

Here, the top 100 most frequent words (with length greater than 4 characters) are selected for visualization.

The scatter plot displays words positioned based on their semantic similarity‚Äîcloser points represent more related words, revealing clusters of similar meaning in the embedding space.

### üîç Visualizing Filtered GloVe Embeddings with t-SNE

This example loads GloVe embeddings, filters out common stopwords and very short words to focus on meaningful terms, and limits the set to the top 200 words.

Next, it applies t-SNE to reduce the high-dimensional vectors to 2D for visualization.

The scatter plot displays word embeddings, with annotations for a subset of 100 words to illustrate their relative positions in semantic space.


```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Load GloVe embeddings with filtering
glove_file = "glove.6B.100d.txt"

def load_glove_embeddings(file_path, max_words=200):
    embeddings = {}
    with open(file_path, encoding="utf-8") as f:
        for line in f:
            values = line.split()
            word = values[0]
            # Filter stopwords and short words
            if word in stop_words or len(word) < 5:
                continue
            if len(embeddings) >= max_words:
                break
            vector = np.array(values[1:], dtype="float32")
            embeddings[word] = vector
    return embeddings

glove_embeddings = load_glove_embeddings(glove_file)

# Prepare data for t-SNE
words = list(glove_embeddings.keys())
vectors = np.array([glove_embeddings[w] for w in words])

# Apply t-SNE to reduce to 2D
tsne = TSNE(n_components=2, random_state=42, perplexity=40, n_iter=3000)
vectors_2d = tsne.fit_transform(vectors)

# Plot the embeddings
plt.figure(figsize=(14, 10))
plt.scatter(vectors_2d[:, 0], vectors_2d[:, 1], s=5, alpha=0.7)

# Annotate a subset to avoid clutter
for i, word in enumerate(words[:100]):
    plt.annotate(word, xy=(vectors_2d[i, 0], vectors_2d[i, 1]), fontsize=9)

plt.title("t-SNE Visualization of GloVe Embeddings (Filtered)")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.show()
```

# Summary of Key NLP Concepts and Techniques

**Natural Language Processing (NLP)** enables machines to understand, interpret, and generate human language. To work effectively with text data, NLP relies on various methods to convert words and documents into numerical forms called **embeddings**.

### 1. Traditional Text Representation Methods

* **Bag of Words (BoW):** Represents documents as counts of unique words, ignoring grammar and word order. Simple but treats all words equally.
* **TF-IDF (Term Frequency-Inverse Document Frequency):** Enhances BoW by weighting words according to how important they are in a document relative to the whole corpus. It downweights common, less informative words.

### 2. Word Embedding Techniques

* **Word2Vec:** Learns dense vector representations of words capturing semantic relationships, via models like CBOW and Skip-Gram. Similar words have similar vectors.
* **FastText:** Extends Word2Vec by representing words as collections of character n-grams, improving embeddings for rare and out-of-vocabulary words.
* **GloVe:** Combines global word co-occurrence statistics with local context windows to learn embeddings capturing both global and local semantics.

### 3. Modern Contextual Embeddings

* Traditional embeddings assign a fixed vector per word regardless of context.
* Modern approaches like **transformers (BERT, GPT, etc.)** produce contextual embeddings that vary with sentence meaning, capturing nuanced language understanding.
* Transformers have become state-of-the-art for complex NLP tasks, though traditional methods still serve well in many practical applications.

### 4. Practical Use Cases and Datasets

* Classic embeddings are effective for tasks like spam detection, text classification, and recommendation systems, especially with limited resources.
* Transformers excel at complex tasks such as question answering, named entity recognition, summarization, and translation.
* Example datasets for hands-on projects include:
  * **Sentiment Analysis:** [Sentiment140](https://www.kaggle.com/datasets/kazanova/sentiment140), [Twitter US Airline Sentiment](https://www.kaggle.com/datasets/crowdflower/twitter-airline-sentiment)
  * **Spam Detection:** [SMS Spam Collection](https://www.kaggle.com/datasets/uciml/sms-spam-collection-dataset), [Enron Email Dataset](https://www.kaggle.com/datasets/mohinurabdurahimova/maildataset)
