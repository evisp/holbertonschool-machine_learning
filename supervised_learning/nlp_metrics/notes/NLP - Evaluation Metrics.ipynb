# Natural Language Processing - Evaluation Metrics

## What are the applications of natural language processing?

Natural Language Processing (**NLP**) is a field of artificial intelligence that **focuses on the interaction between computers and human language**. It involves the development of algorithms and models to enable machines to understand, interpret, and generate human language in a way that is both meaningful and contextually relevant.

Applications of NLP include:

1. **Machine Translation**: Translating text from one language to another, like Google Translate.


2. **Sentiment Analysis**: Determining the emotional tone of a piece of text, often used for social media monitoring and customer feedback analysis.


3. **Speech Recognition**: Converting spoken language into written text, used in virtual assistants like Siri and Alexa.


4. **Named Entity Recognition (NER)**: Identifying and classifying entities such as names of people, places, and organizations in text.


5. **Text Summarization**, **Text Classification** and **Text Generation**: 
- Generating concise summaries of long documents, helpful for news articles and research papers.
- Categorizing text into predefined classes, used for spam filtering and content categorization.
- Creating human-like text, used for various applications, including creative writing and content generation.

6. **Chatbots and Virtual Assistants**: Creating conversational agents that can interact with users in natural language.

NLP has extensive applications across various industries, including healthcare, finance, customer service, education, and more. It continues to advance rapidly, improving the ways in which computers interact with and understand human language.


## What is a BLEU score?

A **BLEU (Bilingual Evaluation Understudy) score is a widely used metric to evaluate the quality of machine-generated text, particularly in the context of machine translation**. 

It **measures the similarity between the generated text and one or more reference (human-generated) texts**. BLEU is designed to assess the **precision** of the generated text by comparing its n-gram (sequence of n words) overlap with the reference texts.

Here's how BLEU works:

1. **N-gram Precision**: BLEU computes precision for different n-gram lengths (usually 1 to 4). Precision for a specific n-gram length is the ratio of the number of matching n-grams in the generated text to the total number of n-grams in the generated text.


2. **Brevity Penalty**: BLEU also includes a brevity penalty to discourage overly short generated texts that might artificially inflate the precision scores. This ensures that longer generated texts are not unfairly penalized.


3. **Geometric Mean**: The individual n-gram precisions are geometrically averaged to compute the overall BLEU score.


#### Example:

Suppose you're evaluating a machine translation system that translates English sentences into French. You have a set of reference translations for a particular input sentence:

**Reference Translations**:

- `Reference 1`: "The cat is on the mat."

- `Reference 2`: "The cat is lying on the rug."


**Machine-Generated Translation**:

- `Machine Translation`: "The cat is on the rug."


**N-grams**:
Let's consider unigrams (individual words) for calculating the BLEU score.

**Precision**:
We calculate the precision of unigrams in the machine-generated translation compared to the reference translations.

- Unigram "`The`" appears 1 time in the reference translations and 1 time in the machine translation.


- Unigram "`cat`" appears 1 time in the reference translations and 1 time in the machine translation.


- Unigram "`is`" appears 2 times in the reference translations and 1 time in the machine translation.


- Unigram "`on`" appears 1 time in the reference translations and 1 time in the machine translation.


- Unigram "`the`" appears 2 times in the reference translations and 0 times in the machine translation.


- Unigram "`mat`" appears 1 time in the reference translations and 0 times in the machine translation.


- Unigram "`lying`" appears 1 time in the reference translations and 0 times in the machine translation.


- Unigram "`rug`" appears 1 time in the reference translations and 1 time in the machine translation.


*Total matches = 6 (count of unigrams present in both reference and machine translation)*


`Precision = Total matches / Total unigrams in machine translation = 6 / 9 = 0.6667 (rounded to 4 decimal places)`


**Brevity Penalty**:

The brevity penalty accounts for differences in the lengths of the machine translation and the reference translations. In this case, the length of the machine translation is 6, and the closest reference length is 7.

`Brevity Penalty = exp(1 - (7 / 6)) ≈ 0.857`


**BLEU Score Calculation**:


`BLEU Score = Brevity Penalty * (Geometric mean of n-gram precisions)`


`BLEU Score = 0.857 * sqrt(0.6667) ≈ 0.750`


In this example, the BLEU score is approximately 0.750, indicating a reasonable match between the machine-generated translation and the reference translations.


Calculating BLEU scores involves importing libraries and utilizing built-in functions. The nltk library (Natural Language Toolkit) provides a convenient way to calculate BLEU scores. Make sure you have nltk installe


```python
import nltk
from nltk.translate.bleu_score import sentence_bleu

# Reference sentences
reference_list = [
    ["The", "cat", "is", "on", "the", "mat."],
    ["The", "cat", "is", "lying", "on", "the", "rug."]
]

# Generated sentence
generated_sentence = ["The", "cat", "is", "on", "the", "mat."]

# Calculate BLEU score
bleu_score = sentence_bleu(reference_list, generated_sentence)

print("BLEU Score:", bleu_score)
```

## What is a ROUGE score?

A **ROUGE (Recall-Oriented Understudy for Gisting Evaluation) score is a set of evaluation metrics commonly used to assess the quality of machine-generated text**, especially in the context of text summarization and natural language generation. **ROUGE measures the similarity between the generated text and one or more reference (human-generated) texts**.

**ROUGE scores primarily focus on capturing recall**, which means how well the generated text covers the content present in the reference texts. Different variants of ROUGE exist, such as ROUGE-N, ROUGE-L, ROUGE-W, and more, each emphasizing different aspects of content overlap.

Here are a few common ROUGE variants:

- **ROUGE-N**: Measures the overlap of n-grams (sequences of n words) between the generated and reference texts. ROUGE-1 considers unigrams (single words), ROUGE-2 considers bigrams, and so on.


- **ROUGE-L**: Computes the longest common subsequence (LCS) between the generated and reference texts. It takes into account the order of words and captures the longest shared sequence.


- **ROUGE-W**: Considers weighted F1-score of overlapping n-grams, giving different weights to different n-grams. This accounts for the fact that longer n-grams are usually more informative.


#### Example:

Suppose you're evaluating an automatic text summarization system. The reference summary is: "`The cat is lying on the rug`." And the generated summary is: "`A cat rests on the rug.`"

For ROUGE-1, you break down the sentences into unigrams (individual words):

- Reference: `["The", "cat", "is", "lying", "on", "the", "rug."]`


- Generated: `["A", "cat", "rests", "on", "the", "rug."]`


In this case, the recall (overlap) of unigrams is 5 out of 7, so ROUGE-1 recall is `5/7 ≈ 0.714`.


For ROUGE-L, you calculate the longest common subsequence (LCS) between the reference and generated sentences. The LCS here is: `["cat", "on", "the", "rug."]`, and its length is 4. ROUGE-L recall is `4/7 ≈ 0.571`.


ROUGE scores provide insights into how well the generated text captures the content of the reference texts. Similar to BLEU, ROUGE has its limitations and is often used alongside other metrics for a more comprehensive evaluation of machine-generated text.






```python
from rouge import Rouge

# Reference and generated summaries
reference_summary = "The cat is lying on the rug."
generated_summary = "A cat rests on the rug."

# Initialize the ROUGE evaluator
rouge = Rouge()

# Calculate ROUGE-1, ROUGE-2, and ROUGE-L scores
scores = rouge.get_scores(generated_summary, reference_summary)

# Print the scores
print("ROUGE-1:", scores[0]['rouge-1']['f'])
print("ROUGE-2:", scores[0]['rouge-2']['f'])
print("ROUGE-L:", scores[0]['rouge-l']['f'])
```

## What is perplexity?

Perplexity is a metric used to evaluate the **quality of language models**, particularly in the field of natural language processing (NLP). It measures **how well a language model predicts a given sequence of words**. In essence, perplexity quantifies how surprised a model is when it encounters a new sequence of words.

A lower perplexity indicates a better-performing language model. A language model with lower perplexity is better at predicting the next word in a sequence, suggesting that it has a better understanding of the language's underlying patterns.

Mathematically, perplexity is calculated as the inverse probability of the test set normalized by the number of words:

$$Perplexity = 2^{- \frac{1}{N} \cdot log_2(likelihood)}$$

Where:

- `N`: Number of words in the test set


- `Likelihood`: Probability of the test set according to the language model


In simpler terms, perplexity can be thought of as the average number of choices the model has to make when predicting the next word in a sequence. Lower perplexity indicates that the model's predictions align well with the actual data, making it less "perplexed" by the data.

## When should you use one evaluation metric over another?

Choosing the right evaluation metric depends on the specific goals of your NLP task and the characteristics of the generated text you're evaluating. Different metrics capture different aspects of language quality and performance. Here's a general guideline for choosing between common NLP evaluation metrics:

1. **BLEU Score**:

- Use when evaluating machine translation or text generation tasks.


- Advantages: Measures n-gram overlap, which helps in assessing the correctness of generated content compared to references.


- Limitations: Doesn't capture fluency, coherence, and meaning; can favor simple and repetitive outputs.



2. **ROUGE Score**:

- Use when evaluating text summarization or content generation tasks.


- Advantages: Measures recall and captures content overlap, useful for tasks where the generated text might not be an exact match to the references.


- Limitations: Ignores fluency and doesn't penalize overly long or repetitive outputs.



3. **Perplexity**:

- Use when evaluating language models.


- Advantages: Reflects the model's ability to predict the next word in a sequence; lower perplexity indicates better predictive performance.


- Limitations: Focuses solely on likelihood; doesn't capture higher-level language quality aspects like semantics and coherence.



4. **Human Evaluation**:

- Use when assessing subjective aspects of generated text.


- Advantages: Provides insights into fluency, coherence, creativity, and overall quality as perceived by humans.


- Limitations: Can be time-consuming, subjective, and resource-intensive; requires human annotators and careful design.


5. **Multiple Metrics**:

- Often, a combination of metrics provides a more comprehensive evaluation.


- Using multiple metrics helps balance trade-offs and get a holistic view of the system's performance.


- Consider using metrics that complement each other to cover different aspects.
