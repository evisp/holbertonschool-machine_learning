# Exploring Data Through Visualization: Matplotlib

**Data visualization** is a powerful method to *understand* and *communicate* complex information in a more accessible way. 


**Matplotlib**, a popular Python library, plays a crucial role in creating various types of plots and charts to reveal patterns and trends within datasets. 


In this tutorial, we will explore different types of plots offered by Matplotlib, ranging from simple line plots to more advanced visualizations like histograms and scatter plots.

### Matplotlib Key Features

[Matplotlib](https://matplotlib.org/) is a popular and widely-used Python library for creating visualizations. Its key features empower users to convey data insights effectively. Here's a concise summary:

- **User-Friendly Interface**: Matplotlib provides an intuitive and user-friendly interface for creating a wide range of plots and charts. Its simplicity makes it accessible for users at various skill levels.

- **Diverse Plot Types**: The library supports a diverse set of plot types, including line plots, bar charts, scatter plots, histograms, and more. 

- **Customization Options**: Matplotlib offers extensive customization options to tailor the appearance of plots. Users can control aspects such as colors, markers, labels, titles, and axes, ensuring the visualizations meet specific requirements.

- **Publication-Quality Graphics**: With Matplotlib, users can generate publication-quality graphics suitable for professional presentations and reports. The library provides high-resolution output and supports export to various file formats.

- **Integration with `NumPy` and `Pandas`**: Matplotlib integrates with `NumPy` and `Pandas`, other popular Python library for numerical operations. 

- **Interactive Plotting**: Matplotlib supports interactive plotting, enabling users to explore and interact with their data dynamically. This feature is particularly useful for data exploration and analysis.

### Matplotlib Most Popular Plots

Matplotlib provides a variety of plot types, each suitable for different types of data representation. Here's a brief summary of key plot types in Matplotlib:

#### Line Plot:

- Purpose: Display the relationship between two continuous variables over a continuous interval.
- Usage: `plt.plot(x, y)`

#### Bar Chart:

- Purpose: Compare values across different categories or display the distribution of a single categorical variable.
- Usage: `plt.bar(categories, values)`

#### Scatter Plot:

- Purpose: Visualize the relationship between two continuous variables in a set of data points.
- Usage: `plt.scatter(x, y)`

#### Histogram:

- Purpose: Show the distribution of a single continuous variable by dividing it into bins.
- Usage: `plt.hist(data, bins)`

#### Pie Chart:

- Purpose: Represent parts of a whole, showing the contribution of each category to the total.
- Usage: `plt.pie(sizes, labels)`

#### Other types of plots such as 
- **Box Plot** and **violin** plot (Box-and-Whisker Plot) to combine aspects of box plots and kernel density plots to show the distribution of data.
- **Heatmap** to isualize the intensity of a phenomenon over a 2D grid, often used for correlation matrices or image data.Violin Plot:
- **Area Plot**  to display the cumulative contribution of different variables over a continuous interval.
- ... and others

### Let's get some data

The [Iris](https://www.kaggle.com/datasets/uciml/iris) dataset is a classic dataset used in machine learning and statistics. 


It consists of measurements from three different species of **iris flowers**: *setosa, versicolor*, and *virginica*. 

For each flower, the dataset includes four features (all measured in centimeters)
- sepal length
- sepal width
- petal length
- petal width 

The goal is often to classify iris flowers based on these measurements. The simplicity and clarity of the Iris dataset make it a popular choice for introductory exercises in data analysis and machine learning.


```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn.datasets import load_iris

iris = load_iris()

# Extract the features and target variable
X = iris.data    # Features (sepal length, sepal width, petal length, petal width)
y = iris.target  # Target variable (species)

# Display the first few rows of the dataset
print("First few rows of the Iris dataset:")
print(X[:5])
print("Target variable:")
print(y[:5])
```

### Let's Explore the Data (no Visualization first)

#### Class Distribution:

Count the number of occurrences of each species.


```python
class_distribution = {species: sum(y == species) for species in set(y)}
print("Class Distribution:")
print(class_distribution)
```

#### Average Feature Values by Species:

Calculate the average values of each feature for each species.

```python
unique_species = set(y)

avg_feature_values = {}

for species in unique_species:
    species_rows = X[y == species]
    avg_values = np.mean(species_rows, axis=0)
    
    avg_feature_values[species] = avg_values.tolist()

print("Average Feature Values by Species:")
for key, value in avg_feature_values.items():
    print(f"{key}: ", end="")
    for el in value:
        print(f"{el:.2f}: ", end="")
    print()
```

### Scatter Plot of Sepal Length vs Sepal Width

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Sepal length vs Sepal width")
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

### Petal Length vs. Petal Width

```python
plt.scatter(X[:, 2], X[:, 3], c=y, cmap='viridis', edgecolors='k', alpha=0.8)
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs. Petal Width')
plt.show()
```

### Histogram of Petal Length


```python
plt.hist(X[:, 2], bins=20, color='skyblue', edgecolor='pink')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title("Petal Length Frequency", fontsize=16)
plt.xlabel('Petal length', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

### Sepal Length Histogram

```python
plt.hist(X[:, 0], bins=20, color='lightcoral', edgecolor='darkred')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title("Sepal Length Frequency", fontsize=16)
plt.xlabel('Sepal length', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

### Distribution of Flower Species in Iris Dataset

```python
# Count the occurrences of each flower species
species_counts = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}
species_distribution = [sum(y == species) for species in species_counts.keys()]

# Plot a pie chart
plt.pie(species_distribution, labels=species_counts.values(), autopct='%1.1f%%', colors=['lightcoral', 'lightgreen', 'lightskyblue'])
plt.title('Distribution of Flower Species in Iris Dataset')
plt.show()
```

### Box Plot: visualizes the distribution of a dataset and identifies outliers

```python
# Box plot for petal length by species
plt.boxplot([X[y == i, 2] for i in range(3)], labels=['Setosa', 'Versicolor', 'Virginica'])
plt.title('Box Plot of Petal Length by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

### Violin Plot

Combines aspects of box plot and kernel density estimation.

```python
import seaborn as sns

# Violin plot for sepal width by species
sns.violinplot(x=y, y=X[:, 1], palette='muted')
plt.title('Violin Plot of Sepal Width by Species')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

### Bar Chart



```python
# Bar chart for average petal width by species
avg_petal_widths = [X[y == i, 3].mean() for i in range(3)]
plt.bar(['Setosa', 'Versicolor', 'Virginica'], avg_petal_widths, color=['lightcoral', 'lightgreen', 'lightskyblue'])
plt.title('Average Petal Width by Species')
plt.xlabel('Species')
plt.ylabel('Average Petal Width (cm)')
plt.show()
```

### Heatmap

```python
import seaborn as sns

# Correlation matrix heatmap
correlation_matrix = np.corrcoef(X, rowvar=False)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', xticklabels=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
plt.title('Correlation Heatmap')
plt.show()
```

### Variation in Petal Width

Let's create a line chart to visualize the variation in petal width for the Setosa species across different data points

```python
import matplotlib.pyplot as plt

# Extracting petal width for the Setosa species
setosa_petal_widths = [X[i][3] for i in range(len(y)) if y[i] == 0]  # Index 3 corresponds to petal width

# Plotting a line chart
plt.plot(range(len(setosa_petal_widths)), setosa_petal_widths, marker='o', color='skyblue')
plt.xlabel('Data Points')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Widths for Setosa Iris Flowers')
plt.show()
```
