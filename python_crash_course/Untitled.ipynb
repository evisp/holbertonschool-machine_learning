```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def count_word_frequency_from_file(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()

        words = content.split()
        word_frequency = {}

        for word in words:
            if len(word) > 4:
                clean_word = word.strip('.,!?').lower()

                if clean_word in word_frequency:
                    word_frequency[clean_word] += 1
                else:
                    word_frequency[clean_word] = 1

        return word_frequency

    except FileNotFoundError:
        print(f"File not found: {file_path}")
        return {}

# Read a sentence from a text file
file_path = "file.txt"
result = count_word_frequency_from_file(file_path)

# Generate and display a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(result)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
