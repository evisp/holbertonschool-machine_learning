## Syntax Rules - Grammar is Everywhere

Previously, we sent the computer two instructions and wrote each on a separate line. If we were to put them both on the same line, we'd get an error:

```python
print(0.00 + 6.99) print(4.5 - 3.5)
```

`print(0.00 + 6.99) print(4.5 - 3.5)` resulted in text describing a **syntax** error. This is because all programming languages — Python included — have syntax rules. Each line of instructions has to comply with these rules.
Running into errors is common when we're programming, and we'll constantly learn about errors and how to fix them. 

Collectively, we call all these *instructions* we write **code**. We call each line of instruction a line of code.

When we write code, we *program* the computer to do something. For this reason, we also call the code we write a **computer program (or a program)**.

The code we write serves as input to the computer. We call the result of executing the code output.

### Comments - a Necessary Break when  Coding
The computer executes code from the first line downward and ignores blank lines. Besides blank lines, the computer also ignores any sequence of characters that comes to the right of the `#` symbol. The sequence of characters that follows the `#` symbol is a **code comment**. We can use code comments to add information about our code:

```python
print(1.99)   # App cost for Fruit Ninja Classic
print(698516) # Rating count
print(4.5)    # Average user rating
```

In Python, comments serve multiple purposes

- Comments can be used to explain Python code.

- Comments can be used to make the code more readable.

- Comments can be used to prevent execution when testing code.

```python
# comments can be in one line
print("Nice commenting!")
```

```python
# or
# separated in multiple lines
# for improved readability
print("I see clearer now")
```

Or, you can use a multiline string.

Since Python will ignore string literals that are not assigned to a variable, you can add a multiline string (triple quotes) in your code, and place your comment inside it:

```python
"""
This is a comment
written in
more than just one line
"""
print("Hi, there!")
```

We'll learn how to save values in Python, how to update values, and how to use syntax shortcuts to update values. We use a single equals sign (`=`) to assign labels to variables. Let's see a few examples of how we can do this.

When we run the code `result = 3.98`, the value `3.98` gets saved in the computer memory. The computer memory has many storage locations, and `3.98` gets saved to one particular location.

![variables](figs/1-variables.svg)

The storage location for `3.98` is a **variable**. When we ran the code `result = 3.98`, we stored `3.98` in a variable (storage location) named `result`, so `result` is a variable name.

Using variable names can be a very useful way to keep track of different variables in Python. For example:

```python
# Use object names to keep better track of what's going on in your code!
my_income = 100

tax_rate = 0.1

my_taxes = my_income*tax_rate
```

```python
# Show my taxes!
my_taxes
```

These are the two syntax rules we need to follow when we're naming variables:

1. We must use only letters, numbers, or underscores (we can't use apostrophes, hyphens, spaces, etc.).
2. Variable names can't begin with a number.

![variables](figs/3-variables.svg)

Note that variable names are case sensitive, which means that a variable named `result` is different than a variable named `Result`. **It is common practice to write python variables in all lowercase letters** such as `taxes = 10.5`, and in case of multiple words to seperate them with the `_` symbol, such as `my_taxes = 10.5`

We can update the value stored in a variable. Below, we first store `1.99` in the variable `app_costs`, and then we update `app_costs` to store `6.99` instead.

```python
app_costs = 1.99
print(app_costs)

app_costs = 6.99
print(app_costs)
```

We can also update a variable doing arithmetic operations. There are several syntax shortcuts we can use to update a variable when we're doing arithmetic operations. Below is a table with some syntax shortcuts we can use:

In other words, the variable must already be defined. When we try to update a variable that we haven't defined, we get an error called `NameError`.

In Python, the `=` operator tells us that the value on the right is assigned to the variable on the left. It doesn't tell us anything about equality. We call `=` an assignment operator, and we read code like `x = 5` as "five is assigned to x" or "x is assigned five," not "x equals five."

### String indexing

We know strings are a sequence, which means Python can use indexes to call parts of the sequence. Let's learn how this works.

In Python, we use brackets `[]` after an object to call its index. We should also note that indexing starts at `0` for Python. Let's create a new object called s and then walk through a few examples of indexing.

```python
# Assign s as a string
s = 'Hi there!'
```

```python
#Check
s
```

```python
# Show first element (in this case a letter)
s[0]
```

```python
s[3]
```

We can use a : to perform slicing which grabs everything up to a designated point. For example:

```python
# Grab everything past the first term all the way to the length of s which is len(s)
s[3:]
```

```python
# Grab everything UP TO the 3rd index
s[:3]
```

```python
#Everything
s[:]
```

We can also use negative indexing to go backwards.

```python
# Last letter (one index behind 0 so it loops back around)
s[-1]
```

```python
# Grab everything but the last letter
s[:-1]
```

We can also use index and slice notation to grab elements of a sequence by a specified step size (the default is 1). For instance we can use two colons in a row and then a number specifying the frequency to grab elements. For example:

```python
# Grab everything, but go in steps size of 1
s[::1]
```

```python
# Grab everything, but go in step sizes of 2
s[::2]
```

```python
# We can use this to print a string backwards
s[::-1]
```

### String formatting

String formatting lets you inject items into a string rather than trying to chain items together using commas or string concatenation. As a quick comparison, consider:

```python
player = 'Thomas'
points = 33

'Last night, '+player+' scored '+str(points)+' points.'  # concatenation

f'Last night, {player} scored {points} points.'          # string formatting
```

There are three ways to perform string formatting.

- The oldest method involves placeholders using the modulo `%` character.
- An improved technique uses the `.format()` string method.
- The newest method, introduced with Python 3.6, uses formatted string literals, called `f-strings`.

##### Formatting with the `.format()` method
The syntax is:

```python
'String here {} then also {}'.format('something1','something2')
```

```python
print('This is a string with an {}'.format('insert'))
```

```python
# Inserted objects can be assigned keywords:
print('First Object: {a}, Second Object: {b}, Third Object: {c}'.format(a=1,b='Two',c=12.3))
```

```python
# Inserted objects can be reused, avoiding duplication:
print('A {p} saved is a {p} earned.'.format(p='penny'))
```

##### Formatted String Literals (`f-strings`)
Introduced in Python 3.6, f-strings offer several benefits over the older `.format()` string method described above. 

```python
name = 'Fred'

print(f"He said his name is {name}.")
```

For more information, visit https://docs.python.org/3/reference/lexical_analysis.html#f-strings
