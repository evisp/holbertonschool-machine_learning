```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn.datasets import load_iris

iris = load_iris()

# Extract the features and target variable
X = iris.data    # Features (sepal length (0), sepal width (1), petal length (2), petal width(3))
y = iris.target  # Target variable (species)
```

```python
plt.figure(figsize=(10,6), dpi=120)
plt.scatter(X[:, 2], X[:, 3], c=y)
plt.title("Sepal length vs Sepal width")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")
plt.savefig("iris.png") 
plt.show()
```

```python
plt.hist(X[:, 0], bins=10, color='skyblue', edgecolor='pink')
plt.grid(axis='y', ls='--', color='#FFDAB9', alpha=0.5)
plt.title("Sepal Length Frequency", fontsize=16)
plt.xlabel("Sepal Length ")
plt.ylabel("Frequency")
plt.show()
```

```python
species_counts       = {0: 'Setosa', 1: 'Versicolor', 2: 'Virginica'}
species_distribution = [sum(y == species) for species in species_counts.keys()]

plt.pie(species_distribution, labels=species_counts.values(), autopct='%1.1f%%',
       colors=['lightblue', 'lightgreen', 'lightcoral'])
plt.title("Distribution of Flower Species in Iris Dataset")
plt.show()
```

```python
plt.boxplot(X, labels=['Sepal length', 'Sepal width', 'Petal length', 'Petal width'], 
            patch_artist=True, 
            boxprops = dict(facecolor = "lightblue"))
plt.title("Iris Features")
plt.show()
```

```python
import seaborn as sns

sns.violinplot(x=y, y=X[:, 1], palette='magma')
```

```python
import numpy as np

correlation_matrix = np.corrcoef(X, rowvar=False)
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True,
           xticklabels=['Sepal length', 'Sepal width', 'Petal length', 'Petal width'])
plt.show()
```
